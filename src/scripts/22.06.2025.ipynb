```python
import os
os.environ['HADOOP_CONF_DIR'] = '/etc/hadoop/conf'
os.environ['YARN_CONF_DIR'] = '/etc/hadoop/conf'

import findspark
findspark.init()
findspark.find()

from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .appName("3")\
        .master("yarn")\
        .config("spark.executor.memory", "4g") \
        .config("'spark.executor.cores", "4")\
        .config("'spark.driver.cores", "4")\
        .config("spark.ui.port", "4051")\
        .getOrCreate()
```

```python

from pyspark.sql import functions as F
from pyspark.sql.types import FloatType



# Загрузка всех событий
events = spark.read.parquet('/user/master/data/geo/events') \
    .select("event_id", "event_type", "user_id", "lat", "lon", "date")

# Загрузка данных городов
geo = spark.read.csv("/user/project_7/geo.csv", header=True, sep=";") \
    .select("city", 
            F.col("lat").cast(FloatType()).alias("city_lat"), 
            F.col("lng").cast(FloatType()).alias("city_lng"))

# Перевод координат в радианы
events = events.withColumn("lat_rad", F.radians(F.col("lat"))) \
    .withColumn("lon_rad", F.radians(F.col("lon")))

geo = geo.withColumn("city_lat_rad", F.radians(F.col("city_lat"))) \
    .withColumn("city_lng_rad", F.radians(F.col("city_lng")))

# Кросс-соединение событий с городами
joined = events.crossJoin(geo)

# Расчет расстояния по формуле Haversine
result = joined.withColumn("distance_km",
    2 * 6371 * F.asin(
        F.sqrt(
            F.pow(F.sin((F.col("city_lat_rad") - F.col("lat_rad")) / 2), 2) +
            F.cos(F.col("lat_rad")) * 
            F.cos(F.col("city_lat_rad")) * 
            F.pow(F.sin((F.col("city_lng_rad") - F.col("lon_rad")) / 2), 2)
        )
    )
)

# Находим ближайший город для каждого события
nearest_city = result.groupBy("event_id", "user_id", "date", "event_type") \
    .agg(
        F.first("city").alias("nearest_city"),
        F.min("distance_km").alias("distance_to_city_km")
    )

# Сохранение результата с партиционированием по дате
nearest_city.write \
    .partitionBy("date") \
    .mode("overwrite") \
    .parquet("/user/reydanyk/project_7/stg")

# Пример вывода
nearest_city.show(10, truncate=False)
```

```python
from pyspark.sql import functions as F


# 1. Загрузка данных
events = spark.read.parquet('/user/master/data/geo/events')
cities = spark.read.csv("/user/project_7/geo.csv", header=True, sep=";") \
    .select("city", 
            F.col("lat").cast("float").alias("city_lat"), 
            F.col("lng").cast("float").alias("city_lng"))

# 2. Вычисление ближайшего города для каждого события
result = events.crossJoin(F.broadcast(cities)) \
    .withColumn("distance_km",
        2 * 6371 * F.asin(
                    F.sqrt(
                        F.pow(F.sin(F.radians(F.col("city_lat") - F.col("lat")) / 2), 2) +
                        F.cos(F.radians(F.col("lat"))) * 
                        F.cos(F.radians(F.col("city_lat"))) * 
                        F.pow(F.sin(F.radians(F.col("city_lng") - F.col("lon")) / 2), 2)))) \
    .groupBy("event", "event_type", "date") \
    .agg(
        F.first("city").alias("nearest_city"),
        F.min("distance_km").alias("distance_to_city_km")
    )

# 3. Сохранение результата
result.write \
    .partitionBy("date") \
    .mode("overwrite") \
    .parquet("/user/reydanyk/project_7/stg")

# Пример вывода
result.show(5, truncate=False)
```
